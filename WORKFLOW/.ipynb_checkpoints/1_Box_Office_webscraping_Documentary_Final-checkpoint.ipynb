*Functions for these last couple*

```python
def movie_details(soup, name_detail): 
    obj = soup.find(text = re.compile(name_detail))
    if not obj: 
        return None
    
    next_element = obj.findNext()
    if next_element:
        return next_element.text
    else: 
        return None

```

```python
def time_to_min(runtime): 
    try:
        minutes = int(runtime.split(" ")[0])*60 + \
                    int(runtime.split(" ")[2])
        return minutes
    except: 
        return None
```

```python
def string_to_money(string): 
    try: 
        simple_string = string.replace('$','').replace(',','')
        return int(simple_string)
    except:
        return None
```

```python
def worldwide_gross_elem(soup): 
    try: 
        worldwide_gross_str = soup.find(class_='mojo-performance-summary-table').\
                    find_all('span', class_='money')[2].text
        worldwide_gross = string_to_money(worldwide_gross_str)
        return worldwide_gross
    except: 
        return None
```

```python
#'Documentary\n    \n        Drama\n    \n        War'
def genres_to_genre(string): 
    try: 
        genres = string.replace(' ', '').split('\n\n')
        return genres
    except: 
        return None
```

```python
def get_cast(url_film, cast_position): 
    try: 
        credits_url = url_film[:46] + 'credits'+'/'
        response_credits = requests.get(credits_url)
        soup_credits = BeautifulSoup(response_credits.text, "lxml")
        cast_regex = re.compile(cast_position)
        cast_strings = soup_credits.find_all(text = cast_regex)
        cast_list = []
        for cast_string in cast_strings:
            cast_name = cast_string.find_previous('a').text.replace('\n\n', '')
            cast_list.append(cast_name)

        return cast_list
    except: 
        return None
```

**BUILDING THE PIPELINE**

```python
## Take: (1st part)
##      Domestic & International & Worldwide Gross, Budget, Runtime from individual movie pages (5 features) & 
## Combine with Genre List with following features (columns in sports movies): (2nd part)
##      link_stub', 'rank','title', 'lifetime_gross', 'max_theaters', 'opening', 
##      'open_th','release_date', 'distributor' (9 features)
```

**1st part: get the info on specific movies**

```python
def get_movie_dict(link, base_url = 'https://www.boxofficemojo.com'):
    '''
    From BoxOfficeMojo link stub, request movie html, parse with BeautifulSoup, and
    collect 
        - title
        - domestic gross
        - international gross 
        - worldwide gross 
        - budget
        - runtime 
    Return information as a dictionary.
    '''
    
    #Create full url to scrape
    url = base_url + link
    
    #Request HTML and parse
    response = requests.get(url)
    movie = response.text
    soup = BeautifulSoup(movie,"lxml")

    
    headers = ['movie_title', 'domestic_total_gross', 'international_gross',
               'worldwide_gross', 'budget', 'runtime', 'MPAA', 'Genres', 'Director', 'Writers', 'Producers']
    
    #Get title
    title = soup.find('title').text.split(" - ")[0]

    #Get domestic gross
    domestic_gross_str = soup.find(class_='mojo-performance-summary-table').\
                        find_all('span', class_='money')[0].text
    domestic_gross = string_to_money(domestic_gross_str)
    
    #Get international gross
    international_gross_str = soup.find(class_='mojo-performance-summary-table').\
                        find_all('span', class_='money')[1].text
    international_gross = string_to_money(international_gross_str)

    #Get worldwide gross
    worldwide_gross = worldwide_gross_elem(soup)
    
    #Get budget
    budget_elem = movie_details(soup,'Budget')
    budget = string_to_money(budget_elem)
    
    #Get runtime
    runtime_str = movie_details(soup, 'Run')
    runtime = time_to_min(runtime_str)
    
    #Get MPAA
    mpaa = movie_details(soup, 'MPAA')
    
    #Get Genres 
    genres_string = movie_details(soup, 'Genres')
    genre = genres_to_genre(genres_string)
    
    #Get Director 
    director = get_cast(url, 'Director')
    
    #Get Writers
    writers = get_cast(url, 'Writer')
    
    #Get Producers
    producers = get_cast(url, 'Producer')
    
    #Create movie dictionary and return
    movie_dict = dict(zip(headers, [title, domestic_gross, international_gross, \
                                    worldwide_gross, budget, runtime, mpaa, genre, director, writers, producers]))
    return movie_dict
```

*Now let's get the info on all the movies in the sports genre*

First make a list of the pages we are going to scrape to get URL's and also some other movie info. 

```python
base_to_base = "https://www.boxofficemojo.com/genre/sg2245914881/?offset="

base_url_list = []
for i in range(20): 
    base_url_list.append(base_to_base + str(i*100))
    
base_url_list
```

```python
movies_all = {}
iteration = 0 
for base_url in base_url_list:
    iteration += 1 
    print(iteration)
    iter_str = int(iteration)
    
    response = requests.get(base_url)
    print(base_url)
    page = response.text

    soup = BeautifulSoup(page,"lxml")
    table = soup.find('table')
    rows = [row for row in table.find_all('tr')]
    
    for row in rows[1:100]:
        items = row.find_all('td')
        link = items[1].find('a')
        title, url = link.text, link['href']
        print(title, url)
        movies_all[title] = [url] + [i.text for i in items]

print(movies_all)
```

```python
docu_movies_all = pd.DataFrame(movies_all).T  #transpose
docu_movies_all.columns = ['link_stub', 'rank','title', 'lifetime_gross', 'max_theaters', 'opening', 'open_th', \
                    'release_date', 'distributor']


docu_movies_all.info()
```

```python
docu_movies_all_page_info_list = []

for link in docu_movies_all.link_stub:
    docu_movies_all_page_info_list.append(get_movie_dict(link))
```

```python
docu_movies_all_page_info = pd.DataFrame(docu_movies_all_page_info_list)  
docu_movies_all_page_info.set_index('movie_title', inplace=True)

docu_movies_all_page_info
```

**2nd part: get the tables**

Now let's get the tables for movies in the sports genre into one dataframe

```python
docu_movies_all = docu_movies_all.merge(docu_movies_all_page_info, left_index=True, right_index=True)

docu_movies_all
```
